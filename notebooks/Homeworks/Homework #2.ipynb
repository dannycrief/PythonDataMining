```python
 import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../../data/country_vaccinations.csv')
```

```python
df.fillna(0, inplace=True)
X = df.drop('source_website', axis=1)
```

```python
X = X[['daily_vaccinations','daily_vaccinations_per_million']]
X = X.rename(columns={"daily_vaccinations": "daily",
                 "daily_vaccinations_per_million": "daily/1M"})
```

```python
# The following heatmap show a static correlation between daily vaccination and daily vaccination per million population
sns.set_style("whitegrid")
ax = plt.axes()
sns.heatmap(X.corr(),ax=ax, linewidths=0.5,
            vmax=1,square=True,cmap="flare",
            linecolor='w',annot=True,annot_kws={"size":10},
            cbar_kws={"shrink": 1})
plt.title('Static correlation between daily vaccination and daily vaccination per million population',
             fontsize=14)
plt.show()
```
